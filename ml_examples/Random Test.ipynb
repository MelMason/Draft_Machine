```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../NCAA_data.csv")
```

```python
df.replace(to_replace ="Star", 
                 value ="Player", inplace=True) 
df.replace(to_replace ="Above Avg", 
                 value ="Player", inplace=True)
df.replace(to_replace ="Below Avg", 
                 value ="Player", inplace=True)

df.head(15)
```

```python
#import label encoder to get numerical values for koi_disposition
from sklearn.preprocessing import LabelEncoder
college = df["college"]
labels = LabelEncoder()
labels.fit(college)
encoded_college = labels.transform(college)
```

```python
encoded_college[:25]
```

```python
df["college_code"] = encoded_college
```

```python
X = df[['win_shares_per_40_minutes', 'field_goal_percentage','games_played',
       'points_per_40','turnovers_per_40', 
        'true_shooting_percentage', 'steals_per_40']]
```

```python
y = df["class"]
```

```python
labels = LabelEncoder()
labels.fit(y)
encoded_y = labels.transform(y)
```

```python
labels_df = pd.DataFrame({"Class": y, "Code": encoded_y}).drop_duplicates(subset=None, keep="first")
labels_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, random_state=42)
```

```python
print("X-train samples",len(X_train))
print("X-test samples",len(X_test))
```

```python
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#import random forest model
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100)
model = model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

```python
feature_names = X.columns
#check feature importance 
sorted(zip(model.feature_importances_, feature_names), reverse=True)
```

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
```

```python
# Train the model with GridSearch
# Create the parameter grid based on the results of random search 
param_grid = {
    'max_depth': [5, 10, 50, 100],
    'n_estimators': [100, 200, 300, 500]
}
# Create a based model
model2 = RandomForestClassifier()
# Instantiate the grid search model
grid = GridSearchCV(model2, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
#check best params and scores
print(grid.best_params_)
print(grid.best_score_)
```

```python
#define model using best hyperparameters
model_opt =  RandomForestClassifier(n_estimators = 200, max_depth = 50)
model_opt.fit(X_train_scaled, y_train)
```

```python
#make predictions with optimized model
predictions2 = model_opt.predict(X_test_scaled)
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions2,
                            target_names=["Player", "Bust"]))
```

```python
#check testing data score 
print(f"Training Data Score: {model_opt.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model_opt.score(X_test_scaled, y_test)}")
```

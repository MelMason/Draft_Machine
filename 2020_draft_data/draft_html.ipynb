```python
X = df[["field_goal_percentage", 'free_throw_attempt_rate', 'height', 'three_point_percentage', 'true_shooting_percentage',
       'win_shares_per_40_minutes']]
```

```python
model = pickle.load(open('../random_forest.pkl', 'rb'))
```

```python
prediction = model.predict(X)
prediction
```

```python
forecast = [prediction[x] for x in range(len(prediction)) ]
```

```python
forecast_string = []
for x in range(len(forecast)):
    if forecast[x] == 1:
        result = "Pro Player"
    else:
        result = "Bust"
    forecast_string.append(result)
```

```python
X["forecast"] = forecast_string
```
